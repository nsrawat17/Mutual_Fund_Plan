## Creating a Mutual Fund Plan with Python
A mutual fund plan is created by selecting the stocks where an investor can benefit in the long term investment.Below I mention the process for creating  a mutual fund plan:
Steps 
1 : Gather historical stock data, such as closing prices and growth trends over time.
2 : Calculate key metrics like Return on Investment (ROI) and volatility (risk) to understand how each stock has performed historically.
3 : Choose stocks that have a high ROI and low volatility to ensure a balance between risk and reward.
4 : Calculate the future value of monthly investments based on the expected ROI of the selected stocks.

# Now check the stock price trend of all companies

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))

for company in df.columns[1:]:
    plt.plot(df['Date'],df[company], label=company , alpha=0.5)

plt.title("Stock Price  Trends of All Indian Companies ", fontsize=16)
plt.xlabel('Date',fontsize = 12)
plt.ylabel('Closing Price (INR)', fontsize=16)


plt.xticks(rotation=45)


plt.legend(loc='center left', bbox_to_anchor=(1,0.5), fontsize=9,title="Companies")

plt.tight_layout()

plt.show()
```

check for which company having highest risk of investing 

```python
all_companies=df.columns[1:]
volatility_all_companies = df[all_companies].std()
volatility_all_companies.sort_values(ascending=False).head(10)
```

Now Check for Company which have highest growth rate 

```python
growth_all_companies =df[all_companies].pct_change()*100
average_growth_all_companies=growth_all_companies.mean()
average_growth_all_companies.sort_values(ascending=False).head(10)
```

Now check for which company give the highest return on investment 

```python
initial_prices_all=df[all_companies].iloc[0]
final_prices_all =df[all_companies].iloc[-1]

roi_all_companies =((final_prices_all-initial_prices_all)/initial_prices_all)*100

roi_all_companies.sort_values(ascending=False).head(10)
```

We want to create a Mutual Fund Plan based on High ROI(Return of investment) and low Risk 

for Selecting companies with high ROI and Low Risk , we can use a combination of ROI and Volatility (Standard Deviation) metrics .
Our main goal is to find companies that give a High Return on Investment  

1. Define ROI and Volatility Threshold 
2. Rank Companies by ROI and Volatility 
3. Assign Investment Ratios

```python
roi_threshold =roi_all_companies.median()
volatility_threshold=volatility_all_companies.median()

selected_companies = roi_all_companies[(roi_all_companies > roi_threshold) & (volatility_all_companies < volatility_threshold)]
selected_companies.sort_values(ascending=False)
```

above company meet the criteria of high ROI and low volatility

To balance the investment between  these compaines , we can use inverse volatility ratio for  allocation.
Companies with lower  volatility will get  a higher weight.

Now Calculate the Weight 

```python
selected_volatility= volatility_all_companies[selected_companies.index]
inverse_volatility= 1/selected_volatility

investment_ratios =(inverse_volatility / inverse_volatility.sum())*100

investment_ratios.sort_values(ascending = False)
```

Analyze our Mutual Fund Plan 

We have created a mutual fund plan for long term investment. Now analyze  and compare our mutual fund plan  by comparing it with the high performing in the stock markert 
Comparing the risk in our mututal fund with the risk in the high growth compaines.

```python
import plotly.graph_objs as go
import plotly.express as px
```

```python
top_growth_compaines = average_growth_all_companies.sort_values(ascending=False).head(10)
risk_growth_rate_companies =volatility_all_companies[top_growth_compaines.index]
risk_mutual_fund_compaines =volatility_all_companies [selected_companies.index]
```

```python
plt.figure(figsize=(10, 6))

# Mutual Fund Companies
plt.barh(
    y=risk_mutual_fund_compaines.index,
    width=risk_mutual_fund_compaines.values,
    color='blue',
    label='Mutual Fund Companies'
)

# Growth Rate Companies
plt.barh(
    y=risk_growth_rate_companies.index,
    width=risk_growth_rate_companies.values,
    color='green',
    alpha=0.7,
    label='Growth Rate Companies'
)

# Labels and Title
plt.xlabel('Volatility (Standard Deviation)')
plt.ylabel('Companies')
plt.title('Risk Comparison: Mutual Fund vs Growth Rate Companies')
plt.legend()

plt.tight_layout()
plt.show()
```

# Compare the ROI of both groups 

```python
expected_roi_mutual_fund= roi_all_companies[selected_companies.index]
expected_roi_growth_companies= roi_all_companies[top_growth_compaines.index]

plt.figure(figsize=(10,6))

plt.barh(
    y=expected_roi_mutual_fund.index,
    width=expected_roi_mutual_fund.values,
    color='blue',
    label='Mutual Fund Compaines'
)

# Bar:  Growth Rate Companies

plt.barh(
    y=expected_roi_growth_companies.index,
    width=expected_roi_growth_companies.values,
    color='green',
    alpha=0.7,
    label='Growth Rate Companies'
)

# ----- Label & Title ----
plt.title('Expected ROI Comparison: Mutual Fund vs Growth Rate Companies')
plt.xlabel('Expected ROI (%)')
plt.ylabel('Companies')
plt.legend(title='Company Type')

plt.tight_layout()
plt.show()
```

The Comparison between the risk(volatility) and expected  ROI  for mutual fund comapines(blue) and growth rate companies shows a clear  trade off. Mutual fund companies offer lower volatility, meaning they are less risky, but also provide lower expected returns.
In contrast, growth rate companies demonstrate higher volatility, indicating more risk, but they offer much higher potential returns, especially companies like Bajaj Auto and Bajaj Finserv.
This highlights a common investment dilemma: lower risk comes with a lower reward, while higher risk could yield higher returns.

For long-term investments, the goal is typically to find companies that offer a balance of stable returns and manageable risk.
The companies in our mutual fund exhibit low volatility, meaning they are less risky, and their moderate returns make them solid choices for long-term, stable growth.
They are well-suited for conservative investors who want steady returns without significant fluctuations in value.

# Calculating Expected  Returns 

```python
import numpy as np 
```

```python
monthly_investment = 5000
years =[1,3,5,10]
n=12    
```

```python
avg_roi= expected_roi_mutual_fund.mean()/100
```

```python
def  future_value(P,r,n,t):
    return P*(((1+r/n)**(n*t)-1)/(r/n))*(1+r/n)

future_values=[future_value(monthly_investment,avg_roi,n,t) for t in years ]


future_value(5000,0.10,12,1)
```

```python
plt.figure(figsize=(10,6))
plt.plot([str(year)+"year" for year in years],future_values,marker ="o",color='red', label='future_value')

plt.title("Expected  value of  Investment of  5000 ruppe per month (mutual Funds)")
plt.xlabel("Investment Period")
plt.ylabel("Future Value(INR)")
plt.grid(True,linestyle='--',color='lightgrey')
plt.legend()

plt.tight_layout()
plt.show()
```

After 1 year, the accumulated value is around â‚¹62,000, and by 5 years, it grows to over â‚¹300,000. The long-term benefit is evident, with the investment growing to nearly â‚¹860,000 over 10 years, which emphasises the value of consistent investing and compounding over time for long-term investors.
